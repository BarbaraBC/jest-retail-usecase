```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read data
features=pd.read_csv('datasets/features.csv')
sales=pd.read_csv('datasets/sales.csv')
stores=pd.read_csv('datasets/stores.csv')
```

```python
# Set date type
features['Date'] = pd.to_datetime(features['Date'])
sales['Date'] = pd.to_datetime(sales['Date'])
```

The features and sales datasets have common fields: Store, Date, IsHoliday. Could they be used to merge the datasets?

```python
print("First Sales record:", sales[0:1].Date)
print("Last Sales record: ",sales[-1:].Date)

print("First Features record: ",features[0:1].Date)
print("Last Features record: ",features[-1:].Date)
```

```python
# Merge datasets
df=pd.merge(sales,features, on=['Store','Date', 'IsHoliday'], how='left')
df=pd.merge(df,stores, on=['Store'], how='left')
df=df.fillna(0)
df['Temperature'] = (df['Temperature']- 32) * 5./9.

types_encoded, types =df['Type'].factorize()
df['Type'] = types_encoded

df.head()
```
